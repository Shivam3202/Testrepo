```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch HTML content from MacroTrends
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Locate the revenue table
table = soup.find("table", {"class": "historical_data_table table"})
rows = table.find_all("tr")

# Extract data into a list
revenue_data = []
for row in rows[1:]:  # Skip the header
    cols = row.find_all("td")
    if len(cols) == 2:  # Ensure it has the expected structure
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()
        revenue_data.append([date, revenue])

# Convert to a DataFrame
tesla_revenue_df = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])
print(tesla_revenue_df.head())
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fetch HTML content from MacroTrends
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
response = requests.get(url)

# Parse HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Locate the revenue table
table = soup.find("table", {"class": "historical_data_table table"})
rows = table.find_all("tr")

# Extract data into a list
revenue_data = []
for row in rows[1:]:  # Skip the header
    cols = row.find_all("td")
    if len(cols) == 2:  # Ensure it has the expected structure
        date = cols[0].text.strip()
        revenue = cols[1].text.strip()
        revenue_data.append([date, revenue])

# Convert to a DataFrame
gamestop_revenue_df = pd.DataFrame(revenue_data, columns=["Date", "Revenue"])
print(gamestop_revenue_df.head())
```

```python
import matplotlib.pyplot as plt

# Plot Tesla stock data
plt.figure(figsize=(12, 6))
plt.plot(tesla_data.index, tesla_data["Close"], label="Tesla Stock Price", color="blue")
plt.title("Tesla Stock Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# Plot Tesla revenue data
plt.figure(figsize=(12, 6))
tesla_revenue_df["Date"] = pd.to_datetime(tesla_revenue_df["Date"])
tesla_revenue_df["Revenue"] = tesla_revenue_df["Revenue"].str.replace(",", "").str.replace("$", "").astype(float)
plt.bar(tesla_revenue_df["Date"], tesla_revenue_df["Revenue"], color="green", label="Tesla Revenue")
plt.title("Tesla Quarterly Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue (in USD)")
plt.legend()
plt.grid(axis="y")
plt.show()
```

```python
# Plot GameStop stock data
plt.figure(figsize=(12, 6))
plt.plot(gamestop_data.index, gamestop_data["Close"], label="GameStop Stock Price", color="red")
plt.title("GameStop Stock Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# Plot GameStop revenue data
plt.figure(figsize=(12, 6))
gamestop_revenue_df["Date"] = pd.to_datetime(gamestop_revenue_df["Date"])
gamestop_revenue_df["Revenue"] = gamestop_revenue_df["Revenue"].str.replace(",", "").str.replace("$", "").astype(float)
plt.bar(gamestop_revenue_df["Date"], gamestop_revenue_df["Revenue"], color="purple", label="GameStop Revenue")
plt.title("GameStop Quarterly Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue (in USD)")
plt.legend()
plt.grid(axis="y")
plt.show()
```
